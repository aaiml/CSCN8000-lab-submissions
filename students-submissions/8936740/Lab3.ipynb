# Part A

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df_cc = pd.read_csv("./creditcard.csv")
#df_cc.head()
```

```python
df_cc.describe()
```

```python
# check data there's any null
print(df_cc.isnull().sum().max()) # max() for finding if there is not 0
```

```python
X = df_cc.drop('Class', axis=1)
y = df_cc['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression(random_state=42)
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
print(f'Accuracy score: {round((accuracy_score(y_test, y_pred) * 100), 2)}%')
```

# Part B

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df_i = pd.read_csv("./insurance.csv")
df_i.head()
```

```python
#change the string to num
sex_mapping = {'female': 1, 'male': 2}
df_i['sex'] = df_i['sex'].map(sex_mapping)
smoker_mapping = {'yes': 1, 'no': 2}
df_i['smoker'] = df_i['smoker'].map(smoker_mapping)
region_mapping = {'southwest': 1, 'southeast': 2, 'northwest': 3, 'northeast': 4}
df_i['region'] = df_i['region'].map(region_mapping)
df_i.head()
```

```python
df_i.describe()
```

```python
# check data there's any null
print(df_i.isnull().sum().max()) # max() for finding if there is not 0
```

```python
X = df_i.drop('charges', axis=1)
y = df_i['charges']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(f'Mean Squared Error: {mean_squared_error(y_test, y_pred)}')
print(f'R-squared: {round(r2_score(y_test, y_pred) * 100, 2)}%')
```
