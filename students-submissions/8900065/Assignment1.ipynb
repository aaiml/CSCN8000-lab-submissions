```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
p_data = pd.read_csv("K:\OneDrive\Desktop\Artificial Lab\pd_speech_features.csv")
p_data
```

```python
p_data.columns = range(p_data.shape[1])
p_data
```

```python
p_data.columns = p_data.iloc[0]
p_data
```

```python
p_data = p_data[1:]
p_data
```

```python
p_data.describe()
```

```python
p_data['class'].value_counts()
```

```python
X = p_data.drop(columns='class', axis=1)
y = p_data['class']

scaler = StandardScaler()
scaler.fit(X)
```

```python
s_p_data = scaler.transform(X)
print(s_p_data)
```

```python
X = s_p_data
y = p_data['class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state = 42)
X.shape
```

```python
X_train.shape
X_test.shape
```

```python
classifier = {
    "SVM = ": SVC(),
    "KNN = ": KNeighborsClassifier(),
    "random_forest = ": RandomForestClassifier(),
}
for name ,model in classifier.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy=accuracy_score(y_test, y_pred)
    print(name, accuracy)
```
