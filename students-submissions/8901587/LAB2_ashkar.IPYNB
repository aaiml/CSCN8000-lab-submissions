```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
from scipy.stats import zscore
```

```python
m_data = pd.read_csv("../../practical_labs/Lab2_dataset.csv")
print(m_data.head())

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(m_data["text"]).toarray()  # Vectorize string values for use in classifier
y = m_data["label_num"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Train and evaluate Gaussian Naive Bayes classifier
# Create an instance of Gaussian Naive Bayes classifier
g_nb = GaussianNB()#is a class in scikit-learn's naive_bayes module that represents the Gaussian Naive Bayes classifier.

# To train the Gaussian Naive Bayes classifier on the training data, you can use the fit() method
g_nb.fit(X_train, y_train)

# Make predictions on the test data using the trained model
g_nb_p = g_nb.predict(X_test)

# Calculate the accuracy of the predictions by comparing with the true labels
g_a = accuracy_score(y_test, g_nb_p)

# Train and evaluate Multinomial Naive Bayes classifier
multinomial_nb = MultinomialNB()#is a class in scikit-learn's naive_bayes module that represents the multinomimal Naive Bayes classifier.
multinomial_nb.fit(X_train, y_train)#use fit for trainig!!!!!same as discussed in the gauusian classifer..
m_nb_p = multinomial_nb.predict(X_test)#make prediction using the train model
m_a = accuracy_score(y_test, m_nb_p)#calculate the accuracy prediction

# print the accuracy !!
print("the accuracy of gaussian clalssiifier is :")
print(g_a)
print("the accuracy if multinomial classifier is :")
print(m_a)
#stop!
```

We get the  gaussian and multinomial accuracy from this program and thats are 1.0 and 0.88(almost 0.9) respectively.As we know that the clasifer that have the highest accuracy perfom better on this given wine dataset than the other .
Finally it is clear that the gaussian classifier is fit for this data set becuase of the highest accuracy ....

```python
d = pd.read_csv("../../practical_labs/AB_NYC_2019.csv")
d.head()
```

Here, we will examine the columns price, minimum_nights, number_of_reviews, and availability_365 to identify any outliers. To start, we will eliminate all the missing values (NA) from the dataset.






```python
cols = ["price", "minimum_nights", "number_of_reviews", "availability_365"]

d[cols] = d[cols].dropna()
cols = ["price", "minimum_nights", "number_of_reviews", "availability_365"]

d[cols] = d[cols].dropna()
z_s = np.abs(zscore(d[cols]))
z_s

```

```python
o_r_z = np.where(z_s > 3)[0]
o_r_z
```

removing outliers 

```python
bclean = d.drop(o_r_z)
bclean.head()
```

Another approach is to replace outliers with mean, median, or mode values. In this case, we will utilize the mean value to replace the outliers.

```python
outlier_rows_indicator = z_s > 3

d_clean = d
d_clean[cols] = np.where(
    outlier_rows_indicator, d[cols].mean(), d[cols]
)

d_clean.head()
```
