```python
# Loading and creating dataframes
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_curve
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

### Loading the Creditcard.csv File

```python
CC = pd.read_csv("C:\\Users\\pdhru\\Downloads\\creditcard.csv")
CC.head()
```

### Creating a model

```python
# Step 1: Preprocessing the data
# Missing values
print(CC.isnull().sum())

# Split the datasets
A = CC.drop('Class', axis=1)
B = CC['Class']

# Scaleing the numerical features
scale = StandardScaler()
A[['Time', 'Amount']] = scale.fit_transform(A[['Time', 'Amount']])

# Step 2: Data Splitting
X_train, X_test, y_train, y_test = train_test_split(A, B, test_size=0.3, random_state=42)

# Step 3: Model Training
M = LogisticRegression()
M.fit(X_train, y_train)

# Step 4: Model Evaluation
y_pred = M.predict(X_test)
Accuracy = accuracy_score(y_test, y_pred)
Precision = precision_score(y_test, y_pred)
Recall = recall_score(y_test, y_pred)
F1 = f1_score(y_test, y_pred)

print(f"Accuracy: {Accuracy}")
print(f"Precision: {Precision}")
print(f"Recall: {Recall}")
print(f"F1-score: {F1}")
```

```python
IN=pd.get_dummies(IN, drop_first = True)
```

```python
A = IN.drop('charges', axis = 1)
B = IN['charges']

X_train, X_test, y_train, y_test = train_test_split(A, B, test_size=0.3, random_state=42)

LR = LinearRegression()
LR.fit(X_train,y_train)
```

### Checking MAE,MSE,R2,RSME

```python
y_pred = LR.predict(X_test)

# Checking the accuracy of the model
MAE = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
RSME = mean_squared_error(y_test, y_pred, squared = False)
R2 = r2_score(y_test, y_pred)

# Print calculate data
print('Mean Absolte: ',round(MAE, 2))
print('Mean Squared: ',round(MSE, 2))
print('Root Mean Squared: ',round(RSME, 2))
print('R-squared: ',round(R2, 2))
```
