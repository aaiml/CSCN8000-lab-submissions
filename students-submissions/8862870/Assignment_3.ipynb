### Part:- 1 Iris Dataset To Check Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

### Loading the iris data

```python
Data = load_iris()
Data
```

### Getting selected data from iris

```python
A = Data.data[:, 2:4]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(A)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

### Making charts

```python
import warnings
warnings.filterwarnings("ignore")

# FInding optimal number of clusters
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plotting the elbow plot
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Plot')
plt.xlabel('Clusters')
plt.ylabel('Inertia')
plt.show()
```

### The 2 differt clusters will be the best for clustering method

```python
from scipy import stats

# Calculateing z-scores
numerical_columns = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
z_scores = stats.zscore(Data2[numerical_columns])

# Define a threshold
threshold = 3

# Find and remove outliers
outliers = Data2[(abs(z_scores) > threshold).any(axis=1)]
df_no_outliers = Data2[~Data2.index.isin(outliers.index)]

print("Original dataset:", Data2.shape)
print("Dataset after removing outliers:", df_no_outliers.shape)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
lb = LabelEncoder()
df_no_outliers['ChestPainType'] = lb.fit_transform(df_no_outliers['ChestPainType'])
df_no_outliers['Sex'] = lb.fit_transform(df_no_outliers['Sex'])
df_no_outliers['RestingECG'] = lb.fit_transform(df_no_outliers['RestingECG'])
df_no_outliers['ExerciseAngina'] = lb.fit_transform(df_no_outliers['ExerciseAngina'])
df_no_outliers['ST_Slope'] = lb.fit_transform(df_no_outliers['ST_Slope'])
df_no_outliers.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
scaler = StandardScaler()

# Applying scaling 
df_scaled = scaler.fit_transform(df_no_outliers)
df_scaled = pd.DataFrame(df_scaled, columns=df_no_outliers.columns)

print(df_scaled.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

A = df_scaled.drop('HeartDisease', axis=1)
B = df_no_outliers['HeartDisease']

# Training and testing sets
X_train, X_test, y_train, y_test = train_test_split(A, B, test_size=0.2, random_state=42)
```

### Making three diffrent models

```python
classifier = {
    "SVM": SVC(),
    "LR": LogisticRegression(),
    "random_forest": RandomForestClassifier(),
}
```

```python
for name ,model in classifier.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy=accuracy_score(y_test, y_pred)
    print(name, accuracy)
```

### The randpom_forest model has the highest accuracy among them at 0.88. The strategic relapse model's accuracy is 0.85, which is the lowest of all of them. The svm model has the second-best accuracy among the others at 0.87.
