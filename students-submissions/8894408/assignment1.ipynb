```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, r2_score
```

```python
# Read dataset
patients_data = pd.read_csv("./pd_speech.csv")
patients_data.head(5)
```

Looking at the data, it seems like the 2nd row (1st body row) is the one that actually represents the column names. We can change to use this as column names.

```python
patients_data.columns = patients_data.iloc[0,]
patients_data.drop(index=0, inplace=True)
patients_data.head(5)
```

#### Data Analysis

```python
print(patients_data.info())
print(patients_data.shape)
```

The dataset consist of 756 samples with 755 features

```python
patients_data.isnull().sum()
```

No null values are found in any of the columns

```python
print(patients_data.duplicated().sum())
```

1 duplicated sample found.

```python
patients_data.describe()
```

#### Data Preprocessing

```python
# Remove duplicates
patients_data.drop_duplicates(inplace=True)
```

#### Data Standardization

```python
# Feature scaling

# scale all numeric columns in the frame.
# "id" gender" and "class" are categorical data and should be excluded from scaling.
non_numeric_columns = ["gender", "class", "id"]
numeric_columns = [
    col for col in patients_data.columns if col not in non_numeric_columns
]

scaler = StandardScaler()
patients_data[numeric_columns] = scaler.fit_transform(patients_data[numeric_columns])
patients_data.head(5)
```

#### Train Model

```python
# Prepare data

# Assuming column "class" indicates the presence or absence of the disease.
X = patients_data.drop(columns=["class"])
y = patients_data["class"]

# Split data to train and test chunks.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=4408
)
# test_size=0.2 indicates train[80%]/test[20%] split.

display(X_train.head(5))
display(y_train.head(5))
```

```python
# Train model with train dataset

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

Now that model is trained and ready, we can evaluate it's performance

#### Evaluation

```python
predictions = model.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, predictions)}")
print(f"R2 score: {r2_score(y_test, predictions, )}")
print(f"Precision: {precision_score(y_test, predictions, pos_label='1')}")
print(f"Recall: {recall_score(y_test, predictions, pos_label='1')}")
```

The model exhibits good perfomance in all the metrics. With higher accuracy, r2 score, precision and moderately high recall.

#### Prediction System

```python
def predict_parkinson(patient_data):
    prediction = model.predict(patient_data)
    if prediction == "1":
        print("Patient is likely to be suffering from Parkinson")
    else:
        print("Patient is unlikely to be suffering from Parkinson")


# Prediction for 5th person in dataframe.
patient_data = patients_data[4:5]
original_label = patient_data["class"]
patient_data = patient_data.drop(columns=["class"])


print(f"Original label: {original_label}")
predict_parkinson(patient_data)
```

As we can see, the original label for sample at 5th position was 1 indicating that person is suffering from Parkinson. Our model correctly predicted the person to likely have the disease.
