```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    mean_absolute_error,
    r2_score,
)
```

#### PART A

```python
cc_data = pd.read_csv("./creditcard.csv")

# Show head of the dataset
cc_data.head()
```

```python
# Show all features in the dataset
cc_data.columns
```

```python
# Data cleaning
cc_data.drop_duplicates(inplace=True)  # Remove duplicates
cc_data.dropna()  # Drop NaN values
```

```python
# Describe each feature with different statistical metrics
display(cc_data.describe())
```

```python
# Transform dataset to training samples and targets
y = cc_data["Class"]  # copy "Class column"
X = cc_data.loc[:, cc_data.columns != "Class"]  # copy all columns except "Class" column
print(y.head())
print(X.head())
```

```python
# Split data to train and test
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=1,
)
model = LogisticRegression(max_iter=1000)
# Train model
model.fit(X_train, y_train)
```

```python
# Perform prediction
y_pred = model.predict(X_test)
```

```python
# Evaluate model performance by finding accuracy and confusion matrix

print(f"Accuracy: {accuracy_score(y_test, y_pred)}")

# Create a confusion matrix
confusion_matrix(y_test, y_pred)
```

#### PART B

```python
# Read insurance dataset
insurance_data = pd.read_csv("./insurance.csv")
```

```python
# Inspect dataset
insurance_data.head()
```

```python
# Data cleaning
insurance_data.drop_duplicates(inplace=True)  # Remove duplicates
insurance_data.dropna()  # Drop NaN values
```

```python
# Transform dataset to training samples and targets
y = insurance_data["charges"]  # copy "Class column"
X = insurance_data.loc[
    :, insurance_data.columns != "charges"
]  # copy all columns except "Class" column
print(y.head())
print(X.head())
```

```python
# Regression only accepts numerical features for training. Convert categorical features to numberic to use in regression.
categorical_columns = ["sex", "smoker", "region"]
label_encoder = LabelEncoder()
for cat_col in categorical_columns:
    X[cat_col] = label_encoder.fit_transform(
        X[cat_col]
    )  # apply encoding for each column

X.head()
```

```python
# Split to train/test datasets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# Train model and perform predictions
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# Evaluation
# Since the target is continous value, accuracy is measured wrt metrics like R2 score, MAE etc.
print(f"R-squared (R2) Score: {r2_score(y_test, y_pred)}")
print(f"Mean Absolute Error (MAE): {mean_absolute_error(y_test, y_pred)}")
```
