#### Problem Statement: Consider the dataset Credit Card Fraud Detection from Kaggle and build a machine-learning model that detects whether a credit card transaction is fraudulent. Demonstrate the steps of data preprocessing and analysis, consider applying train (0.7) and test (0.3), using the logistic regression to build the model, and evaluate to determine the accuracy.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, r2_score, mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

#### Load the dataset

```python
data = pd.read_csv('C:/Users/Arin/Downloads/creditcarddata/creditcard.csv')
data.head(10)
```

#### Explore the dataset

```python
print(data.shape)
print(data.info())
print(data['Class'].value_counts())
```

#### Data preprocessing

```python
# Checking Duplicate rows
print("Number of duplicate rows:", data.duplicated().sum())

# Remove duplicate rows
data.drop_duplicates(inplace=True)

# Reset the index after removing duplicates
data.reset_index(drop=True, inplace=True)

print("Number of duplicate rows after cleaning:", data.duplicated().sum())

# Handle missing values (if any)
data.dropna(inplace=True)

print('Number of normal transaction', len(data[data.Class == 0]))
print('Number of fraud transaction', len(data[data.Class == 1]))

# Separate features (X) and target variable (y)
X = data.drop('Class', axis=1)
y = data['Class']

# Scale the features (except the 'Class' column)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#### Split the dataset into a training set (70%) and a testing set (30%)

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42)
```

#### Train the model

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

#### Predict the model

```python
y_pred = log_reg.predict(X_test)
```

#### Evaluate the model

```python
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", classification_rep)
```

#### Conclusion

##### The model achieves a very high accuracy of 99.92%. However, when dealing with imbalanced datasets like credit card fraud detection, where non-fraudulent transactions outnumber fraudulent ones, accuracy alone may not be sufficient.

##### The confusion matrix sheds light on the types of errors the model is making. Specifically, there were 60 fraudulent transactions classified as non-fraudulent (false negatives), and 11 non-fraudulent transactions classified as fraudulent (false positives).

#### Problem Statement: Use the following insurance dataset and build a predictive system to predict insurance costs. Demonstrate the steps of data preprocessing and analysis, consider applying train (0.7) and test (0.3), using linear regression to build the model, and evaluate the accuracy of predicting the insurance cost.

#### Load the dataset

```python
# Load the dataset
insurance_data = pd.read_csv('C:/Users/Arin/Downloads/creditcarddata/insurance.csv')
insurance_data.head(10)
```

#### Explore the dataset

```python
print(insurance_data.shape)
print(insurance_data.info())
print(insurance_data.describe())
```

#### Data preprocessing

```python
print("Number of duplicate rows:", insurance_data.duplicated().sum())

# Remove duplicate rows
insurance_data.drop_duplicates(inplace=True)

# Reset the index after removing duplicates
insurance_data.reset_index(drop=True, inplace=True)

print("Number of duplicate rows after cleaning:",
      insurance_data.duplicated().sum())

# Handle missing values (if any)
insurance_data.dropna(inplace=True)

# Convert categorical variables into numerical using one-hot encoding (if necessary)

categorical_columns = ['sex', 'smoker','region']
# Now changing catergorical variables into numeric variables
label_encoder = LabelEncoder()
for cat_col in categorical_columns:
    insurance_data[cat_col] = label_encoder.fit_transform(
        insurance_data[cat_col])

display(insurance_data)

# Separate features (X) and target variable (y)
X = insurance_data.drop('charges', axis=1)
y = insurance_data['charges']
```

#### Split the dataset into a training set (70%) and a testing set (30%)

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)
```

#### Train the model

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

#### Predict the model

```python
y_pred = lin_reg.predict(X_test)
```

#### Evaluate the model

```python
accuracy = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
# Print the evaluation metrics
print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R2) Score: {accuracy}")
```

#### Conclusion
##### A lower MSE value indicates that the model's predictions are closer to the actual values.
##### A higher R-squared value indicates a better fit of the model to the data and better predictive performance.
