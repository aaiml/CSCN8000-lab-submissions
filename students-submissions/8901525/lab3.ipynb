## loading the creditcard

```python
df = pd.read_csv('creditcard.csv')
df.head()
```

## making model

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler

# Step 1: Load the dataset

# Step 2: Data Preprocessing
# Check for missing values
print(df.isnull().sum())

# Split the dataset into features (X) and target variable (y)
X = df.drop('Class', axis=1)
y = df['Class']

# Scale numerical features (except 'Time' and 'Amount') to have zero mean and unit variance
scaler = StandardScaler()
X[['Time', 'Amount']] = scaler.fit_transform(X[['Time', 'Amount']])

# Step 3: Data Splitting
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Model Training
model = LogisticRegression()
model.fit(X_train, y_train)

# Step 5: Model Evaluation
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```

```python
insurance=pd.get_dummies(insurance, drop_first = True)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error


X = insurance.drop('charges', axis = 1)
y = insurance['charges']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

linreg = LinearRegression()
linreg.fit(X_train,y_train)

```

## checking the MAE,MSE,R2,RSME

```python
y_pred = linreg.predict(X_test)
# Check the accuracy of model
MAE = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
RSME = mean_squared_error(y_test, y_pred, squared = False)
R2 = r2_score(y_test, y_pred)

# Print calcutaed data
print('Mean Absolte Error is: ',round(MAE, 2))
print('Mean Squared Error is: ',round(MSE, 2))
print('Root Mean Squared Error is: ',round(RSME, 2))
print('R-squared(r2) is: ',round(R2, 2))
```
