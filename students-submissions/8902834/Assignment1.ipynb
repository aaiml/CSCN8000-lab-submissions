### Problem Statement: Given the symptoms exhibited in a patient and a database of anonymized patient records, predict whether the patient is likely to have an illness (Parkinson’s disease). A model of this decision problem could be used by a program to provide decision support to medical professionals.

### Importing the necessary packages:

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### Loading the 'pd_speech_features.csv' dataset in a dataframe:

```python
dataFrame = pd.read_csv("./pd_speech_features.csv")
dataFrame.head(10)
```

### Data Analysis:

```python
dataFrame.info()
```

- Checking for missing value:

```python
print(dataFrame.isnull().sum())
```

- Checking for duplicate data:

```python
print(f"The number of duplicate data: {dataFrame.duplicated().sum()}")
```

### After performing Data Analysis we can conlude:
- The data used in this study were gathered from 188 patients with PD (107 men and 81 women) with ages ranging from 33 to 87 (65.1Â±10.9) at the Department of Neurology in CerrahpaÅŸa Faculty of Medicine, Istanbul University. 
- The control group consists of 64 healthy individuals (23 men and 41 women) with ages varying between 41 and 82 (61.1Â±8.9).

### Data Preprocessing:

- As we can see, the first row of the data are feature names. Therefore we will change the coloumn name to feature names and drop the first row from the data:

```python
dataFrame.columns = dataFrame.iloc[0]
dataFrame.drop(index = 0, inplace = True)
dataFrame.head()
```

- Removing the duplicate data:

```python
dataFrame = dataFrame.drop_duplicates()
print(f"The number of duplicate data: {dataFrame.duplicated().sum()}")
```

### Data Standardization:

- Removing the 'gender' and 'class' column before scaling so that their values are not affected:

```python
gender = dataFrame['gender']
classData = dataFrame['class']
newDataFrame = dataFrame.drop(columns=['gender', 'class'])
newDataFrame.head(10)
```

- Standardizing the remaining data:

```python
scaler = StandardScaler()
standardData = pd.DataFrame(scaler.fit_transform(newDataFrame), columns=newDataFrame.columns)
```

- Adding the 'gender' and 'class' columns to the standardized data:

```python
standardData['gender'] = gender.values
standardData['class'] = classData.values
standardData.head(10)
```

### Model Training:

- Splitting the data into training dataset(80%) and testing dataset(20%):

```python
X = standardData.drop(columns=['class'])
y = standardData['class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=34)
```

- Training the model using Logistic Regression:

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

### Measuring Model's accuracy:

```python
accuracy = accuracy_score(y_test, model.predict(X_test))
print(f"Accuracy score of the Logistic Regression Model: {accuracy * 100} %")
```

- Build a predictive system to verify that the model built is able to predict the patient’s Parkinson’s disease with one instance of the dataset:

```python
n = X_test.shape[0]
randomindex = np.random.randint(0, n)
result = model.predict(pd.DataFrame(X_test.iloc[randomindex, :]).T)
if result == 0:
    print("The model prediction for 1st instance:", result)
    print("Person does not have Parkinson's disease.")
else:
    print("The model prediction for 1st instance:", result)
    print("The person has Parkinson's disease.")
```
