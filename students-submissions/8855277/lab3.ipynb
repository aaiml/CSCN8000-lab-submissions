**Part A:**

Problem Statement: Consider the dataset Credit Card Fraud Detection from Kaggle and build a machine-learning model that detects whether a credit card transaction is fraudulent. Demonstrate the steps of data preprocessing and analysis, consider applying train (0.7) and test (0.3), using the logistic regression to build the model, and evaluate to determine the accuracy.
https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

### Import dataset

```python
data = pd.read_csv('./creditcard.csv')
data.head(10)
```

```python
data.describe()
```

After taking a look at the data, feature 'Time' and 'Amount' need to be normalized and will be done in preprocessing.

```python
data["Class"].value_counts()
```

After checking the distribution of dataset column 'Class', '0' indicates non-fraud transaction and '1' indicates fraud transaction.

So there are 284,315 non-fraud transactions and 492 fraud transactions in total in the dataset.

```python
# separate features and target variables
X = data.drop('Class', axis=1)
X
```

```python
y = data['Class']
y
```

```python
# split train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# data preprocessing
scaler = StandardScaler()
X_train[['Time', 'Amount']] = scaler.fit_transform(X_train[['Time', 'Amount']])
X_test[['Time', 'Amount']] = scaler.transform(X_test[['Time', 'Amount']])
```

```python
# Build Logistic Regression model
log_reg_model = LogisticRegression(random_state=42)
log_reg_model.fit(X_train, y_train)
```

```python
# make prediction
y_pred = log_reg_model.predict(X_test)
```

```python
# evaluate the model with accuracy score
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
# get confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
# get classification report
repo = classification_report(y_test, y_pred)
print(repo)
```

**Part B:**

Problem Statement: Use the following insurance dataset and build a predictive system to predict insurance costs. Demonstrate the steps of data preprocessing and analysis, consider applying train (0.7) and test (0.3), using linear regression to build the model, and evaluate the accuracy of predicting the insurance cost.
https://www.kaggle.com/datasets/mirichoi0218/insurance 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
df = pd.read_csv('./insurance.csv')
df.head(10)
```

```python
# data preprocessing
# convert categorical data to numeric
df = pd.get_dummies(df, columns=['sex', 'smoker', 'region'], drop_first=True)
df.head(10)
```

```python
# split data into features and target
X = df.drop('charges', axis=1)
y = df['charges']
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# build linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
```

```python
# make prediction
y_pred = model.predict(X_test)
```

```python
# model evaluation
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)

r2 = r2_score(y_test, y_pred)
print('R squared', r2)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', rmse)
```
