```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error, r2_score
```

# Part A
### Problem Statement: Consider the dataset Credit Card Fraud Detection from Kaggle and build a machine-learning model that detects whether a credit card transaction is fraudulent. Demonstrate the steps of data preprocessing and analysis, consider applying train (0.7) and test (0.3), using the logistic regression to build the model, and evaluate to determine the accuracy. https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud

```python
# Load Credit Card Fraud Detection dataset
Credit_card = pd.read_csv('creditcard.csv')
```

```python
# Check few rows of dataset
Credit_card.head()
```

```python
# Check statstical summary of numerical features
Credit_card.describe()
```

```python
# Check missing values in dataset
Credit_card_null = Credit_card.isnull().sum()
Credit_card_null
```

```python
# Split the data
X = Credit_card.drop('Class', axis = 1)
y = Credit_card['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# Train the model
import warnings 
warnings.filterwarnings('ignore')
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Evaluate the model
prediction = model.predict(X_test)
```

```python
# Check the accuracy of model
Accuracy = accuracy_score(y_test, prediction)
print('Accuracy of model is:' , Accuracy)
```

# Part B

### Use the following insurance dataset and build a predictive system to predict insurance costs. Demonstrate the steps of data preprocessing and analysis, consider applying train (0.7) and test (0.3), using linear regression to build the model, and evaluate the accuracy of predicting the insurance cost. https://www.kaggle.com/datasets/mirichoi0218/insurance

```python
# Load Insurance datset
Insurance = pd.read_csv('insurance.csv')
```

```python
# Check few rows of datset
Insurance.head()
```

```python
# Check statstical summary of numerical features
Insurance.describe()
```

```python
# Change categorical variable to numeric
Insurance = pd.get_dummies(Insurance, drop_first = True)
```

```python
# Check missing value in dataset
Insurance_null = Insurance.isnull().sum()
Insurance_null
```

```python
# Split the data
X = Insurance.drop('charges', axis = 1)
y = Insurance['charges']
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# Train the model
Insurance_model = LinearRegression()
Insurance_model.fit(X_train1, y_train1)
```

```python
# Evaluate the model
Insurance_prediction = Insurance_model.predict(X_test1)
```

```python
# Check the accuracy of model
MAE = mean_absolute_error(y_test1, Insurance_prediction)
MSE = mean_squared_error(y_test1, Insurance_prediction)
RSME = mean_squared_error(y_test1, Insurance_prediction, squared = False)
R2 = r2_score(y_test1, Insurance_prediction)

# Print calcutaed data
print('Mean Absolte Error is: ', round(MAE, 2))
print('Mean Squared Error is: ', round(MSE, 2))
print('Root Mean Squared Error is: ', round(RSME, 2))
print('R-squared(r2) is: ', round(R2, 2))
```
