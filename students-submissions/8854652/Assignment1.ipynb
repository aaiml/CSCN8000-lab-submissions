```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# Load data
parkinsons_data = pd.read_csv('pd_speech_features.csv')
# Print the data
parkinsons_data
```

```python
# Remove existing column names as original column names are mentioned in first row
parkinsons_data.columns = range(parkinsons_data.shape[1])
# Print data with removed column names
parkinsons_data
```

```python
# Assign first row data as a new column names
parkinsons_data.columns = parkinsons_data.iloc[0]
# Print data with new column names
parkinsons_data
```

```python
# Remove first as column names are still there in it
parkinsons_data = parkinsons_data[1:]
# Print data with proper format
parkinsons_data
```

```python
# Check description of data in dataframe
parkinsons_data.describe()
```

```python
# Check the value count for class
parkinsons_data['class'].value_counts()
```

```python
# Seperate data and labels
X = parkinsons_data.drop(columns='class', axis=1)
y = parkinsons_data['class']
```

```python
# Standardize the data
parkinsons_data_scaler = StandardScaler()
parkinsons_data_scaler.fit(X)
```

```python
# Transform the data
standardize_parkinsons_data = parkinsons_data_scaler.transform(X)
# Print transformed data in similar range
print(standardize_parkinsons_data)
```

```python
# Define label and target for model prediction
X = standardize_parkinsons_data
y = parkinsons_data['class']
```

```python
# split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state = 42)
```

```python
# Check the shape of X
X.shape
```

```python
# Check the shape of X train
X_train.shape
```

```python
# Check the shape of X test
X_test.shape
```

```python
# Create support vector machine classifier with linear kernel
classifier = svm.SVC(kernel='poly')
```

```python
# Train the classifier
classifier.fit(X_train, y_train)
```

```python
# Prdict the data
X_train_prediction = classifier.predict(X_train)
# Check the accuracy of train data
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
# Print accuracy of train data
print('Accuracy score of the training data for Parkinson model is ', training_data_accuracy)
```

```python
# Prdict the data
X_test_prediction = classifier.predict(X_test)
# Check the accuracy of test data
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
# Print accuracy of test data
print('Accuracy score of the test data for Parkinson model is ', test_data_accuracy)
```
