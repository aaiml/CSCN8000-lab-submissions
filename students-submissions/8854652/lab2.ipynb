```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

## PART A

```python
# Load dataset
data = pd.read_csv('Lab2_dataset.csv')
data.head()
```

```python
X = data['text']
y= data['label_num']
```

```python
# Find the unique words(features) from the X_train data and converts them into a matrix
v = CountVectorizer()
X_count = v.fit_transform(X.values)
X_array = X_count.toarray()
```

```python
# Split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X_array, y, test_size = 0.2, random_state = 42)
```

```python
# Train model using Gaussian classifier
Gaussian_model = GaussianNB()
Gaussian_model.fit(X_train, y_train)
```

```python
# Train model using Multinomial classifier
Multinomial_model = MultinomialNB()
Multinomial_model.fit(X_train, y_train)
```

```python
# Predict test data for Gaussian classifier
Gaussian_prediction = Gaussian_model.predict(X_test)

# Predict test data for Multinomial classifier
Multinomial_prediction = Multinomial_model.predict(X_test)
```

```python
# Calculate accuracy of Gaussian classifier
Gaussian_accuracy = accuracy_score(y_test, Gaussian_prediction)
print('Gaussian classifier accuracy is ', Gaussian_accuracy)

# Calculate accuracy of Multinomial classifier
Multinimial_accuracy = accuracy_score(y_test, Multinomial_prediction)
print('Multinomial classifier accuracy is ', Multinimial_accuracy)
```

Based on accuracy score Multinomial classifier model has a higher accuracy score. Therefore, Multinomial model performs better in given scenario.

```python
# Load dataset
AB_NYC = pd.read_csv('AB_NYC_2019.csv')
AB_NYC
```

```python
AB_NYC.price.describe()
```

```python
# Mean based outlier removal

# Calculate mean of price 
mean_price = AB_NYC['price'].mean()

# Calculate standard deviation of price
std_price = AB_NYC['price'].std()

# Define the threshold range
lower_threshold = mean_price - (3 * std_price)
upper_threshold = mean_price + (3 * std_price)

# Remove outliers
clean_AB_NYC_mean = AB_NYC[(AB_NYC['price'] >= lower_threshold) & (AB_NYC['price'] <= upper_threshold)]

# Show data without outlier based on mean
clean_AB_NYC_mean
```

```python
# Median based outlier removal

# Calculate median
median_price = AB_NYC['price'].median()

# Calculate median absolute deviation (MAD)
mad_price = np.median(np.abs(AB_NYC['price'] - median_price))

# Define the threshold range
lower_threshold = median_price - (3 * mad_price)
upper_threshold = median_price + (3 * mad_price)

# Remove outliers
clean_AB_NYC_median = AB_NYC[(AB_NYC['price'] >= lower_threshold) & (AB_NYC['price'] <= upper_threshold)]

# Show data without outlier based on median
clean_AB_NYC_median
```

```python
# Calculate lower percentiles (1st)
lower_percentile = AB_NYC['price'].quantile(0.01)  

# Calculate upper percentiles (99th)
upper_percentile = AB_NYC['price'].quantile(0.99)  

# Remove outliers
clean_AB_NYC_percentile = AB_NYC[(AB_NYC['price'] >= lower_percentile) & (AB_NYC['price'] <= upper_percentile)]

# Show data without outlier based on percentile
clean_AB_NYC_percentile
```

```python
AB_NYC_upper = AB_NYC[AB_NYC.price > upper_percentile]
AB_NYC_upper
```

```python
# Replace upper percentile values with NAN
AB_NYC_upper['price'] = np.NAN
```

```python
# Replace NaN values with mean
AB_NYC_mean = AB_NYC_upper.fillna(mean_price)

# Show data with changes
AB_NYC_mean
```

```python
# Replace NaN values with median
AB_NYC_median = AB_NYC_upper.fillna(median_price)

# Show data with changes
AB_NYC_median
```
