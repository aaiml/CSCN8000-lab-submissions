```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

- Data Analysis

```python
pSData = pd.read_csv("pd_speech_features.csv", header=1)
pSData.head(10)
```

- Data Preprocessing

```python
# Check for any missing values
print(pSData.isnull().sum())
```

```python
pSData.drop_duplicates(inplace=True)  # Remove duplicates
pSData.dropna()  # Removing NaN values
pSData.head(10)
```

- Data Standardization

```python
X = pSData.drop(columns=['class'])
y = pSData['class']
```

```python
# Standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

- Model Training

-  a. Use 80% of samples as training data size.


```python
# Split the data into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
model = LogisticRegression()
model.fit(X_train, y_train)
```

- Measure your modelâ€™s accuracy.


```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```

- Build a predictive system to verify that the model built is able to predict the patientâ€™s Parkinsonâ€™s disease with one instance of the dataset.

```python
# Assuming 'new_instance' is a single instance from the dataset for prediction
new_instance = X.iloc[0, :].values  # Selecting the first instance from the dataset for prediction

# Preprocess the new instance using the same scaler used for training data
new_instance_scaled = scaler.transform([new_instance])

# Use the trained model to predict whether the patient is likely to have Parkinson's disease
prediction = model.predict(new_instance_scaled)

# 'prediction' will contain the predicted label (0: No Parkinson's, 1: Parkinson's)
print("Predicted label:", prediction[0])
```
